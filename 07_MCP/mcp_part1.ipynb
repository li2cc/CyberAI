# 🔗 1 — The M×N Problem MCP Solves

- Every AI app × every tool = a custom integration (M×N glue)
- MCP makes it M+N: tools speak one standard protocol
- Write a tool once as an MCP server → any MCP client can use it
- Think 'USB-C for AI tools'

> ### 🎤 Instructor Script
>
> Start with the problem. If you have M AI applications and N tools, connecting them ad hoc means M times N custom integrations — a brittle mess. The Model Context Protocol turns that into M plus N: each tool is exposed once as an MCP server speaking a standard protocol, and any MCP client can use it. It's often called USB-C for AI tools. That standardization is why MCP matters for building reusable, swappable security tooling.

# 🧩 2 — Host, Client, Server + Primitives

- **Host**: the app the user interacts with (e.g., an IDE, an agent)
- **Client**: lives in the host, holds one connection to a server
- **Server**: exposes capabilities — **tools**, **resources**, **prompts**
- Tools = actions; Resources = read-only data; Prompts = templates

> ### 🎤 Instructor Script
>
> MCP has three roles. The host is the application the user works in — an IDE or an agent. Inside it, a client maintains a connection to a server. The server exposes capabilities of three kinds: tools, which are actions the model can invoke like a port scan; resources, which are read-only data the model can fetch like a CVE list; and prompts, which are reusable templates. Our security server will expose tools and a resource.

# 🛠️ 3 — Build a Server with FastMCP

- `FastMCP` makes a server from decorated Python functions
- `@mcp.tool()` turns a function into a callable tool
- `@mcp.resource(...)` exposes read-only data
- Our server (mcp_server.py) exposes simulated security tools

> ### 🎤 Instructor Script
>
> Building a server is delightfully simple with FastMCP. You create a server object, then decorate ordinary Python functions: at-mcp-dot-tool turns a function into a tool the client can call, and at-mcp-dot-resource exposes read-only data. Our mcp_server.py, modeled on Omar's cyber_mcp_server, exposes safe simulated tools — a port scan, a CVE lookup, a target list. The cell below shows the shape of that server; the full file is in this folder.

**The shape of `mcp_server.py`** (the real, runnable file is in this folder):

```python
from mcp.server.fastmcp import FastMCP

mcp = FastMCP("cyber-tools")

@mcp.tool()
def port_scan(target: str) -> str:
    """Simulated nmap-style scan of a lab host (safe, offline)."""
    hosts = {"10.0.0.5": "21/ftp 22/ssh 80/http", "10.0.0.8": "445/smb 3306/mysql"}
    return hosts.get(target, "(no open ports)")

@mcp.tool()
def lookup_cve(service: str) -> str:
    return "vsftpd 2.3.4 -> CVE-2011-2523 (backdoor)" if "vsftpd" in service else "no CVE found"

if __name__ == "__main__":
    mcp.run()   # speaks MCP over stdio
```

# 🔌 4 — Transports: stdio vs HTTP

- **stdio**: client launches the server as a subprocess (local, simple)
- **HTTP/SSE**: server runs as a web service (remote, multi-client)
- We use **stdio** — the client starts `mcp_server.py` for us
- Same tools either way; only the connection differs

> ### 🎤 Instructor Script
>
> A server has to talk to its client over some transport. The two common ones are stdio, where the client launches the server as a subprocess and they talk over standard input and output, and HTTP, where the server runs as a web service many clients can reach. We use stdio because it's the simplest for local development — the client starts mcp_server.py automatically. The tools behave identically; only the plumbing changes. In Part 2 we connect a client over stdio and call these tools for real.
