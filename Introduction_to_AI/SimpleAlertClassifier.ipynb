```python
import os
from typing import Optional

def get_api_key(key_name: str) -> Optional[str]:
    """Get API key from environment or Google Colab userdata"""
    # Try environment variable first
    env_key = os.getenv(key_name)
    if env_key:
        return env_key
    
    # Try Google Colab userdata if available
    try:
        from google.colab import userdata
        return userdata.get(key_name)
    except ImportError:
        return None

# Get API keys
anthropic_key = get_api_key('ANTHROPIC_API_KEY')
if not anthropic_key:
    raise ValueError(f"ANTHROPIC_API_KEY not found in environment or Colab secrets")
    
print("✓ API key loaded successfully")
```

```python
from anthropic import Anthropic

client = Anthropic(api_key=anthropic_key)

alert = "5 failed logins for 'admin' from 203.0.113.7 in 20s, then 1 success"
```

```python
msg = client.messages.create(
    model='claude-haiku-4-5-20251001', max_tokens=150,
    messages=[{'role':'user',
      'content': f'Classify benign/suspicious and explain in one sentence: {alert}'}])
print(msg.content[0].text)
```

```python
from openai import OpenAI
import textwrap

openai_key = get_api_key('OPENAI_API_KEY')
if not openai_key:
    raise ValueError("OPENAI_API_KEY not found in environment or Colab secrets")

client_openai = OpenAI(api_key=openai_key)

alert = "5 failed logins for 'admin' from 203.0.113.7 in 20s, then 1 success"

response = client_openai.chat.completions.create(
    model="gpt-4-mini",
    messages=[{"role": "user", "content": f"Classify benign/suspicious and explain in one sentence: {alert}"}]
)

text = response.choices[0].message.content

print("\n".join(textwrap.wrap(text, width=50)))
```
