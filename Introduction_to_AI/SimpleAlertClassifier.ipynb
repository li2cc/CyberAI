```python
from google.colab import userdata
#This is for testing purpose. Never display your key to the public.
userdata.get('ANTHROPIC_API_KEY')
```

```python
from anthropic import Anthropic
client = Anthropic(api_key=userdata.get('ANTHROPIC_API_KEY'))

alert = "5 failed logins for 'admin' from 203.0.113.7 in 20s, then 1 success"



```

```python
msg = client.messages.create(
    model='claude-haiku-4-5-20251001', max_tokens=150,
    messages=[{'role':'user',
      'content': f'Classify benign/suspicious and explain in one sentence: {alert}'}])
print(msg.content[0].text)
```

```python


from openai import OpenAI
from google.colab import userdata
import textwrap

client = OpenAI(api_key=userdata.get("OPENAI_API_KEY"))

alert = "5 failed logins for 'admin' from 203.0.113.7 in 20s, then 1 success"

response = client.responses.create(
    model="gpt-4.1-mini",
    input=f"Classify benign/suspicious and explain in one sentence: {alert}"
)

text = response.output[0].content[0].text

print("\n".join(textwrap.wrap(text, width=50)))
```
